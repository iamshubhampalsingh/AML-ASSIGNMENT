```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]

# Download the dataset and save it as a CSV file
df = pd.read_csv(url, header=None, names=column_names)
df.to_csv("iris_dataset.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset from the "Downloads" folder
file_path = r"C:\Users\51010243\Downloads\iris.csv"  # Replace with your actual file path

# Read the dataset into a Pandas DataFrame
df = pd.read_csv(file_path)

# Scatter plot for Sepal Length vs. Petal Length
plt.figure(figsize=(12, 4))

# a) Sepal Length vs. Petal Length
plt.subplot(131)
plt.scatter(df["sepal_length"], df["petal_length"])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Sepal Length vs. Petal Length")

# b) Petal Width vs. Sepal Width
plt.subplot(132)
plt.scatter(df["petal_width"], df["sepal_width"])
plt.xlabel("Petal Width")
plt.ylabel("Sepal Width")
plt.title("Petal Width vs. Sepal Width")

# c) Petal Length vs. Sepal Length
plt.subplot(133)
plt.scatter(df["petal_length"], df["sepal_length"])
plt.xlabel("Petal Length")
plt.ylabel("Sepal Length")
plt.title("Petal Length vs. Sepal Length")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset from the "Downloads" folder
file_path = r"C:\Users\51010243\Downloads\iris.csv"  # Replace with your actual file path

# Read the dataset into a Pandas DataFrame
df = pd.read_csv(file_path)

# 5-point summary using describe()
summary = df.describe()

# Visualize the 5-point summary with a box plot
plt.figure(figsize=(8, 6))
summary.drop("count").T.plot.box(vert=False)
plt.title("5-Point Summary for Numeric Attributes")
plt.xlabel("Value")
plt.grid(axis="x")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset from the "Downloads" folder
file_path = r"C:\Users\51010243\Downloads\iris.csv"  # Replace with your actual file path

# Read the dataset into a Pandas DataFrame
df = pd.read_csv(file_path)

# Create a histogram for Sepal Length with 5 bins
plt.figure(figsize=(12, 4))

plt.subplot(121)
plt.hist(df["sepal_length"], bins=5, edgecolor='k')
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length (5 bins)")

# Create a histogram for Petal Width with 5 bins
plt.subplot(122)
plt.hist(df["petal_width"], bins=5, edgecolor='k')
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.title("Histogram of Petal Width (5 bins)")

plt.tight_layout()
plt.show()
```
